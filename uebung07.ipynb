```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return np.exp(-x**x) * np.sin(y)

x = np.linspace(-3, 10, 3)
y = np.linspace(-3, 10, 3)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.pcolormesh(X, Y, Z)
plt.show()
```
